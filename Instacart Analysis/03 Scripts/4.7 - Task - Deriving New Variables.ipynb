### If-Statements with User-Defined Functions

```python
# define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

#### 4. Identify busiest hours of the day by 'most orders', 'avg orders' , 'fewest orders'

```python
# identify which hours are busiest and slowest
ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
# creating distribution list in order to identify top, middle, and bottom 
dist_list = list(ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).quantile([0.25, 0.5, 0.75]))
```

```python
#checking output
dist_list
```

```python
# converting above value counts of ords_prods_merged into data dictionary to use frequency count
dict_hours = ords_prods_merged['order_hour_of_day'].value_counts(dropna=False).to_dict()
```

```python
dict_hours
```

```python
# creating list to assign appropriate order level
result3 = []

for value in ords_prods_merged["order_hour_of_day"]:
    for count, orders in dict_hours.items():
        if value == count:
          if orders <= dist_list[0]:
            result3.append("Fewest orders")
          elif orders > dist_list[0] and orders < dist_list[2]:
            result3.append("Average orders")
          elif orders >= dist_list[2]:
            result3.append("Most orders")
```

```python
result3.count("Fewest orders")
```

```python
result3.count("Average orders")
```

```python
result3.count("Most orders")
```

```python
# merge result3 list to ords_prods_merged df
ords_prods_merged['busiest_period_of_day'] = result3
```

```python
# validating frequency in ords_prods_merged df
ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False)
```
