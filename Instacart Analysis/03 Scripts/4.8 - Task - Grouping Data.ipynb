#### 2. Find the aggregated mean of the 'order_number' column grouped by 'department_id'

```python
# creating group by function to group order_number data by department_id
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

#### 4. Creating a loyalty flag for existing customers

```python
# creating a new column 'max_order' to generate number of orders by user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# validate that new column appears in dataset
ords_prods_merge.head()
```

Creating loc for 'loyalty flag'

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# verify totals per flag group
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# re-verifying shape of dataset after loc
ords_prods_merge.shape
```

#### 6. Create a spending flag for each user based on the average price across all orders

1. If the mean of the prices of products purchased by a user is < than 10 = flag as 'Low spender'
2. If the mean of the prices of products purchased by a user is >= to 10 = flag as 'High spender'

```python
# creating a new column 'avg_price' to generate mean price of products by user
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# validate that new column appears in dataset
ords_prods_merge.head(15)
```

Creating loc for 'spending flag'

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# verify totals per flag group
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# re-verifying shape of dataset after loc
ords_prods_merge.shape
```

```python
# re-validate that new columns appears in dataset
ords_prods_merge.head(15)
```

Average price and spending_flag aligned

#### 7. Create 'order frequency flag' that marks regularity of a user's ordering behaviour according to the median in the 'days since prior order' column

1. If the median of 'days since prior order' is > than 20, customer flag = 'non-frequent customer'
2. If the median of 'days since prior order' is > than 10 and <= 20, customer flag = 'regular customer'
3. If the median of 'days since prior order' is <= 10, customer flag = 'frequent customer'

```python
# creating a new column 'median_prior_order' to generate order frequency by user
ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# validate that new column appears in dataset
ords_prods_merge.head(15)
```

Creating loc for 'order_frequency_flag'

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_order'] > 10) & (ords_prods_merge['median_prior_order'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# verify totals per flag group
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
